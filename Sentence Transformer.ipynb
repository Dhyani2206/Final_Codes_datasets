```python
import pandas as pd

excel_file_path = "merged_chemistry_history.csv"
df = pd.read_csv(excel_file_path)
df
```

```python
from sentence_transformers import SentenceTransformer

# Load a pre-trained sentence transformer model
model = SentenceTransformer('paraphrase-mpnet-base-v2')
```

```python
import pandas as pd

excel_file_path = "mergedactual_chemistry_history.csv"
df2 = pd.read_csv(excel_file_path)
df2
```

```python
import torch

def dense_retrieval(query, model, description_embeddings):
    query_embedding = model.encode(query, convert_to_tensor=True)
    # Convert DataFrame to PyTorch tensor
    description_embeddings_tensor = torch.tensor(description_embeddings.values)
    similarities = torch.nn.functional.cosine_similarity(query_embedding.unsqueeze(0), description_embeddings_tensor)
    sorted_indices = similarities.argsort(descending=True)
    return sorted_indices

query = "Baggy clothes"
sorted_indices = dense_retrieval(query, model, df)
for idx in sorted_indices[:10]:
    print(df['0'][idx.item()])
```
